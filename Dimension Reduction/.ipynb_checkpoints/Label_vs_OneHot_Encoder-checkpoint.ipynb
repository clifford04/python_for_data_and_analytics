```python
df = pd.read_csv("diabetes.csv")
df.dropna(inplace=True)
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
one_hot = OneHotEncoder()
one_hot_df = one_hot.fit_transform(df[['gender']]).toarray()
```
