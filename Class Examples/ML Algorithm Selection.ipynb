```python
import pandas as pd
import numpy as np
from itertools import combinations
from itertools import permutations
```

```python
def extract_state(geography):
    arr = geography.split(',')
    return arr[-1].strip()
```

```python
df['State'] = df['Geography'].apply(extract_state)
```

```python
df.drop(['binnedInc','Geography'], axis=1, inplace=True)
df.dropna(inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
def label_encode(df, columns):
    for col in columns:
        le = LabelEncoder()
        col_values_unique = list(df[col].unique())
        le_fitted = le.fit(col_values_unique)
 
        col_values = list(df[col].values)
        #le.classes_
        col_values_transformed = le.transform(col_values)
        df[col] = col_values_transformed
        
```

```python
X = df.drop('TARGET_deathRate', axis=1)
label_encode(X, ['State'])
y = df[['TARGET_deathRate']]
```

```python
X.dtypes
```

```python
def create_column_combinations(num_of_combinations):
    comb_list = []
    for i in range(1, num_of_combinations):
        comb_list = comb_list + list(combinations(X.columns, i))
    
        return comb_list
```

```python
# To split the dataset into train and test datasets
from sklearn.model_selection import train_test_split

from sklearn.neighbors import KNeighborsRegressor

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def train_model(features, target):
    features_train, features_test, target_train, target_test = train_test_split(features, target, test_size = 0.2, random_state = 10)
    model = KNeighborsRegressor(n_neighbors=3)
    model.fit(features_train, target_train)
    score = model.score(features_test, target_test)
    return score

```

```python
accuracy_scores = []
comb_list = create_column_combinations(4)
print(comb_list)
for cols in comb_list:
    cols_list = list(cols)
    features = X[cols_list]
    target = y
    model_score = train_model(features, target)
    temp = [str(cols_list), model_score]
    accuracy_scores.append(temp)
```

```python
accuracy_scores = pd.DataFrame(accuracy_scores, columns=['COL_NAMES','SCORE'])
```
